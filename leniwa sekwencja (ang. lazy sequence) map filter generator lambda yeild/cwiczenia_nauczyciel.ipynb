## Zadanie 3: Podwajanie liczb nieparzystych
Utw贸rz list liczb od 1 do 15. U偶yj `filter`, aby wybra tylko liczby nieparzyste, a nastpnie `map`, aby je podwoi.

```
liczby = range(1, 16)
nieparzyste = filter(lambda x: x % 2 == 1, liczby)
podwojone = list(map(lambda x: x * 2, nieparzyste))
print(podwojone)
```

## Zadanie 4: Generator kwadrat贸w
Napisz funkcj-generator `kwadraty(n)`, kt贸ra zwraca kolejne kwadraty liczb od 0 do n-1. Wypisz wszystkie wartoci.

```
def kwadraty(n):
    for i in range(n):
        yield i * i

for x in kwadraty(5):
    print(x)
```

## Zadanie 5: Liczby z tekstu
Dany jest napis: `'4 7 10 13 20'`. Rozdziel go na liczby, zamie elementy na `int`, a nastpnie wybierz tylko te, kt贸re dziel si przez 5.

```
tekst = '4 7 10 13 20'
liczby = map(int, tekst.split())
podzielne = list(filter(lambda x: x % 5 == 0, liczby))
print(podzielne)
```

## Zadanie 8: Generator liczb parzystych
Napisz generator `parzyste(n)`, kt贸ry zwraca tylko liczby parzyste mniejsze od n.

```
def parzyste(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

for x in parzyste(10):
    print(x)
```

## Zadanie 9: czenie map i filter
Utw贸rz list liczb od 1 do 20. Najpierw u偶yj `filter`, aby wybra tylko liczby podzielne przez 3, a potem `map`, aby zamieni je na ich kwadraty.

```
liczby = range(1, 21)
podzielne3 = filter(lambda x: x % 3 == 0, liczby)
kwadraty = list(map(lambda x: x ** 2, podzielne3))
print(kwadraty)
```

## Zadanie 10: Operacje na tekcie
U偶ytkownik wprowadza tekst. Podziel go na sowa (`split`), a nastpnie za pomoc `map` utw贸rz list dugoci ka偶dego sowa. U偶yj `filter`, aby zostawi tylko sowa du偶sze ni偶 3 znaki.

```
tekst = input('Podaj tekst: ')
slowa = tekst.split()
dlugosci = list(map(len, slowa))
dlugie = list(filter(lambda x: x > 3, dlugosci))
print(dlugosci)
print(dlugie)
```
