### Import plików CSV w Pythonie

**CSV (Comma-Separated Values)** to popularny format plików do przechowywania danych tabelarycznych. Pliki CSV są szeroko używane w analizie danych, ponieważ są proste, czytelne dla człowieka i obsługiwane przez większość narzędzi do analizy danych.

W Pythonie mamy kilka sposobów na wczytywanie plików CSV:
1. Wbudowany moduł `csv`
2. Biblioteka `pandas` (najczęściej używana w analizie danych)
3. Biblioteka `numpy`

```python
import csv
```

#### Metoda 1: Używanie wbudowanego modułu `csv`

```python
# Sprawdź ścieżkę pliku, jeśli plik .csv znajduje się w tym samym katalogu co notebook
# Zmieniam ścieżkę na dostępny przykładowy plik CSV
with open("Datasets/StudentsMarks.csv", "r") as csvfile:
    reader_variable = csv.reader(csvfile, delimiter=",")
    for row in reader_variable:
        print(row)

print(reader_variable)
```

#### Metoda 3: Używanie biblioteki `numpy`

```python
# Wczytanie pliku CSV do tablicy numpy (tylko dane numeryczne)
# np.loadtxt - tylko liczby
# np.genfromtxt - radzi sobie z mieszanymi danymi

import numpy as np

data = np.loadtxt(
    "Datasets/StudentsMarks.csv",
    delimiter=",", #ogranicznik
    skiprows=1, # pomija nagłówek, np.loadtxt() potrafi wczytać tylko wartości numeryczne
    usecols=(0, 3)  # tu wpisz numery kolumn z liczbami
)

print("Kształt danych:", data.shape)
print("Pierwsze 5 wierszy:")
print(data[:5])
```

```python
data = np.genfromtxt(
    "Datasets/StudentsMarks.csv",
    delimiter=",", #ogranicznik
    skip_header=1, # pomija nagłówek, działa tak samo, jak skiprows=1
    dtype=None,    # pozwala na typ mieszany
    encoding="utf-8"
)

print(data[:5])
print(type(data))
```

### Najlepsze praktyki przy wczytywaniu plików CSV

**Kiedy używać której metody:**

- **`csv`**: Gdy potrzebujesz podstawowego wczytania danych bez dodatkowych operacji
- **`pandas`**: Najlepsza opcja do analizy danych - oferuje bogaty zestaw funkcji do manipulacji danymi
- **`numpy`**: Gdy pracujesz głównie z danymi numerycznymi i potrzebujesz wydajnych operacji matematycznych

**Uwagi:**
- Zawsze sprawdź ścieżkę do pliku
- Zwróć uwagę na separator (przecinek, średnik, tabulator)
- Sprawdź kodowanie pliku (UTF-8, CP1250, itp.)
- Obsługuj wyjątki (FileNotFoundError, UnicodeDecodeError)
