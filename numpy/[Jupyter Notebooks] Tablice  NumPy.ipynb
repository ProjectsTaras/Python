- Przykład: Biorąc pod uwagę tablicę NumPy, której podstawowe dane są typu „int32". Zmień typ danych danego obiektu na „float64".

```python
# Utwórz tablicę NumPy
Astypearr1D = np.array([10, 20, 30, 40, 50])
# Wydrukuj tablicę
print(Astypearr1D)
print(Astypearr1D.dtype)
```

```python
# Zmień typ danych na „float64"
Astypearr1D= Astypearr1D.astype('float64')
print(Astypearr1D)
print(Astypearr1D.dtype)
```

### Dołącz dane do tablicy NumPy

```python
# Utwórzmy nową tablicę i spróbuj dołączyć dane do tablicy
AppendArray= np.array([[1,2,3],[4,5,6]]) 
AppendArray
```

```python
# Dołączanie elementów do tablicy (append)
print(np.append(AppendArray,[7,8,9])) 
```

```python
print (np.append(AppendArray, [[7,8,9]],axis = 0))    # elementy wzdłuż osi x
```

```python
print (np.append(AppendArray, [[5,5,5],[7,8,9]],axis = 1)) # elementy wzdłuż osi Y
```

### Usuń dane z tablicy NumPy

```python
# Utwórzmy nową tablicę i spróbuj usunąć dane
Deletearr2D = np.array([[23,45,67],[78,65,92]])
print(Deletearr2D)
```

```python
# Usuwanie elementów z tablicy (delete)
object=[0,1]
Deletearr2D = np.delete(Deletearr2D,object)
Deletearr2D
```

### Wstaw dane do tablicy NumPy

```python
# Utwórzmy nową tablicę i spróbuj wstawić dane
Insertarr2D = np.array([[23,45,67],[78,65,92]])
print(Insertarr2D)
```

```python
# Wstawianie elementów do tablicy (insert)
Insertarr2D = np.insert(Insertarr2D, 1, 9, axis = 1)
Insertarr2D
```

```python
Insertarr2D = np.insert(Insertarr2D, [1,2], 9, axis = 1)
Insertarr2D
```
