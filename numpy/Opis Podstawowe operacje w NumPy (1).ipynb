## Sekcja 1: Operacje arytmetyczne w NumPy
### Metoda i operator dodawania

```python
# Importuj bibliotekę NumPy
import numpy as np
```

```python
# Utwórz dwie tablice tego samego rozmiaru i wymiarów
# Użyj funkcji add()
AddarrA=np.array([10,20,30,40])
AddarrB=np.array([1,2,3,4])
np.add(AddarrA,AddarrB)
```

```python
# NumPy - operator dodawania
AddarrA,AddarrB
```

```python
# Dodaj tablice za pomocą operatora „+”
AddarrA+AddarrB
```

### Metoda i operator odejmowania

```python
# Utwórz dwie tablice o tym samym rozmiarze i wymiarach
# Użyj funkcji subtract()
SubtractarrA=np.array([25,15,35,45])
SubtractarrB=np.array([10,6,11,18])
np.subtract(SubtractarrA,SubtractarrB)
```

```python
# NumPy - operator odejmowania
SubtractarrA,SubtractarrB
```

```python
# Odejmij tablice za pomocą operatora „-”
SubtractarrA-SubtractarrB
```

### Metoda i operator mnożenia

```python
# Utwórz dwie tablice o tym samym rozmiarze i wymiarach
# Użyj funkcji multiply()
MultiplyarrA=np.array([3,5,7,9,11])
MultiplyarrB=np.array([2,4,6,8,10])
np.multiply(MultiplyarrA,MultiplyarrB)
```

```python
# NumPy - operator mnożenia
MultiplyarrA,MultiplyarrB
```

```python
# Pomnóż dwie tablice za pomocą operatora „*”
MultiplyarrA*MultiplyarrB
```

### Metoda i operator dzielenia

```python
# Utwórz dwie tablice o tym samym rozmiarze i wymiarach
# Użyj funkcji divide()
DividearrA=np.array([10,15,27,39,51])
DividearrB=np.array([2,3,9,13,16])
np.divide(DividearrA,DividearrB)
```

```python
# NumPy - operator dzielenia
DividearrA,DividearrB
```

```python
# Podziel tablice za pomocą operatora "/"
DividearrA/DividearrB
```

## Sekcja 2: Operacje macierzowe w NumPy
### Iloczyn skalarny - np.inner i np.dot
### Obie funkcje liczą iloczyn skalarny
Dla tablic jednowymiarowych (wektorów 1D) obie funkcje działają tak samo — wykonują iloczyn dwóch wektorów i sumują wyniki.

```python
# Wektory jako tablice 1D NumPy
# Utwórz dwa wektory
InnerarrA = np.array([1, 2, 3])
InnerarrB = np.array([4, 5, 6])
print(InnerarrA,InnerarrB)
```

```python
# Oblicz iloczyn skalarny
np.inner(InnerarrA,InnerarrB)
```

```python
# Oblicz iloczyn DOT
np.dot(InnerarrA,InnerarrB)
```

### Dla tablic dwuwymiarowych (macierzy 2D) funkcje np.inner i np.dot działają inaczej.
### np.dot wykonuje mnożenie macierzy, a np.inner liczy iloczyny skalarne między wierszami.
- np.dot(A, B) — mnożenie macierzowe (wiersze × kolumny)<br>
Wykonuje standardowe mnożenie macierzy:<br>
wiersze z A mnożone są przez kolumny z B.<br>

- np.inner(A, B) — iloczyn wewnętrzny (wiersz–wiersz)<br>
Funkcja liczy iloczyn skalarny każdego wiersza A z każdym wierszem B.<br>
Można ją stosować tylko wtedy, gdy wiersze A i B mają tę samą długość<br>
(czyli liczba kolumn w A musi być równa liczbie kolumn w B).<br>

```python
# Macierze jako obiekty ndarray
DotarrA = np.array([[1, 2], [3, 4]])
DotarrB = np.array([[5, 6, 7], [8, 9, 10]])
print(DotarrA)
```

```python
print(DotarrB)
```

```python
# Zastosuj iloczyn DOT
np.dot(DotarrA,DotarrB)
```

```python
InnerarrA = [[1, 2],
          [3, 4]]

InnerarrB = [[10, 20],
          [30, 40]]
np.inner(InnerarrA,InnerarrB)
```

## Sekcja 3: Operacje statystyczne w NumPy

### Średnia <br>
Oblicza średnią arytmetyczną elementów.

```python
# Utwórz tablicę
ArrMean = np.array([5,6,7])
ArrMean
```

```python
# Średnia
print(np.mean(ArrMean))
```

### Mediana <br>
Zwraca wartość środkową po posortowaniu danych.

```python
# Utwórz tablicę
ArrMedian=np.array([5,6,7,10,1,100])
ArrMedian
```

KROK 1: Posortowanie danych<br>
KROK 2: Wyznaczenie wartości środkowej. Tablica ma 6 elementów (liczba parzysta), więc mediana to: element → 6 i element → 7<br>
NumPy liczy średnią z tych dwóch: (6+7)/2=6.5<br>

```python
# Mediana
print(np.median(ArrMedian))
```

### Odchylenie standardowe - np.std() <br>
Mierzy rozrzut danych wokół średniej (jak bardzo dane są „rozsiane”).<br><br>
Odchylenie standardowe mówi, jak bardzo wartości są rozrzucone wokół średniej:<br>
małe odchylenie → dane są blisko średniej<br>
duże odchylenie → wartości są bardziej rozproszone<br>
<br>
Krok 1: Średnia<br>

(5+6+7)/3=6<br>
Krok 2: Różnice od średniej<br>

5 − 6 = −1<br>
6 − 6 = 0<br>
7 − 6 = 1<br>
Krok 3: Kwadraty różnic<br>

(−1)² = 1<br>
0² = 0<br>
1² = 1<br>
Krok 4: Średnia z kwadratów<br>

(1+0+1)/3=2/3≈0.6667<br>
Krok 5: Pierwiastek 0.6667 ≈ 0.81649658<br>

Wynik:<br>
np.std(ArrStdDev) = 0.81649658<br>

###Podsumowanie<br>

Funkcja np.std():<br>

✔ znajduje średnią<br>
✔ liczy różnice od średniej<br>
✔ podnosi je do kwadratu<br>
✔ uśrednia<br>
✔ wyciąga pierwiastek<br>

```python
# Utwórz tablicę
ArrStdDev=np.array([5,6,7])
ArrStdDev
```

```python
# Odchylenie standardowe
print(np.std(ArrStdDev))
```
